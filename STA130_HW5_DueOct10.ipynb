# QUESTION 1

### What is the key factor that makes the difference between ideas that can, and cannot be examined and tested statistically?
For it to be possible to test an idea, it must be possible to make that idea a hypothesis that had a variable we can measure. So something like "Does coffee improves grades?" is testable because we can gather information on grades and whether or not each person drinks coffee. We could then look at that data to see if there was any correlation those 2 variables. On the other hand a question like "Does God exist?" cannot be tested for statistically. You can measure how many people believe in God, or their different opinions on the subject, but there's no concrete data that would let us prove or disprove if God exists?

### What would you describe is the key "criteria" defining what a good null hypothesis is?
I think that for a null hypothesis to be considered "good" it must meet the following criteria:
1. The null hypothesis must be testable, meaning that there should be an empirical way for us to confirm that it is more or less likely to be true, or maybe even prove/disprove it entirely. If a null hypothesis isn't testable then it is meaningless in the context of statistics, since it wouldn't provide us with any insight.
2. The good null hypothesis should be relatively specific, because a null hypothesis that is too general doesn't actually accomplish anything either, since proving or disproving it true would still mean that a bunch of different things might still be possible.
3. A null hypothesis should also be "default", where there is no "effect" or "difference" for whatever is being tested. Basically the null hypothesis should assume that there is no change, and then the alternate hypothesis should try to see how strong the evidence is to support it.

### And what is the difference between a null hypothesis and an alternative hypothesis in the context of hypothesis testing?
The null hypothesis is the thing that we assume to be true, unless something proves it otherwise. The alternate hypothesis is whatever the null hypothesis isn't, that is to say that the goal isn't to prove either of the hypotheses. Instead, we are trying to see how strong the evidence is against the null hypothesis. And then if the evidence against the null hypothesis is strong enough, we may choose to reject the null hypothesis and instead accept the alternate hypothesis.

### What $x_i$, $\bar{x}$, $\mu$, and $\mu_0$ mean:
$x_i$: This symbol simply means represents a single person (or data point) from the information we sampled. So if we were looking at average GPA at UofT and we samples a small group of people, then each person's grade would be an $x_i$, because it represents a single data point from our sample.

$\bar{x}$: This symbol represents the average that we get from all the different heights from within our sample. Going back the the UofT GPA analogy, $\bar{x}$ would be the average GPA of all of the people that we talked to.

$\mu$: Since we aren't able to (feasibly) talk to every single student at UofT to find out their GPA, the average GPA of the people that we talked to isn't necessarily the average GPA of everyone at UofT. And in the same way that we represent the average GPA of our sample of the student body with $\bar{x}$, we use the symbol $\mu$ to represent the GPA of *everyone* (i.e. the population) at UofT, even if we aren't always able to find the true value of something like this. 

$\mu_0$: Where $\mu$ is the actual average of the population, $\mu_0$ is the value that we assume $\mu$ to be when formulating the null hypothesis. For example we might assume that $\mu_0$ is 3.2, when in reality $\mu$ is 2.9.

### Connection to the sentence:
When we are trying to test a hypothesis, we are using the data from our sample (so $x_i$ and $\bar{x}$) to see if we can infer something about the population as a whole from it. We are seeing if there is enough evidence to reject that the average GPA = $\mu_0$ and instead assume that an alternative average is the correct one. Since the outcome of the test "refer(s) to the population parameter, rather than the sample statistic", we know that the outcome of our hypothesis testing will refer to $\mu$ instead of $\bar{x}$.

# QUESTION 3

The reason that we "imagine a world where the null hypothesis is true", is because the whole point of hypothesis testing is to figure out how likely it is to get our observed result or more extreme values. So for example we're testing to see if our coin is fair, we say that the null hypothesis is that we get heads 50% of the time and tails 50% of the time. Then if we observe that we actually got tails 60% of the time, we see how likely that is to happen assuming our null hypothesis is true. If the likelyhood of something like that happening is low (assuming our null hypothesis), then we reject the null hypothesis in favor of the alternate hypothesis. So assuming the null hypothesis to be true is essential for proper hypothesis testing.

# QUESTION 4

The p-value is the probability of seeing results as or more extreme than our own, if we assume the null hypothesis to be true. So a smaller p-value would mean that our observations are very unlikely to happen if the null hypothesis were true. So for example if we take our coin flipping experiment and we flip the coin 100 times to get 95 heads and only 5 tails, we might calculate a p-value of 0.0001, that would mean that there is a 0.01% chance of seeing data as extreme as ours given a true null hypothesis. If data is very unlikely to happen, then that might make us question whether the null hypothesis is true. So the smaller the p-value (i.e. the less likely it is for our data the occur given the null hypothesis) the more likely it becomes that the null hypothesis is false.

## Problem Introduction

We are here investigating whether 80 students from STA130 can distinguish between cups of tea where milk is poured first and cups where tea is poured first. This analysis aims to test if STA130 students have a genuine ability to discern the pouring order or if their performance is consistent with random guessing.

### Relationship Between This Experiment and the Original with Fisher and Bristol

The original experiment by Fisher and Bristol involved Dr. Bristol tasting cups of tea and identifying whether the tea or milk was poured first. Fisher designed this test to determine if Dr. Bristol's claims of being able to distinguish the pouring order were statistically significant or if her success could be attributed to chance. The key difference between the original experiment and this one lies in the nature of the participants.

The original experiment focused on Dr. Bristol's ability, which is a specific and potentially more personalized parameter. On the other hand, the STA130 experiment generalizes to a broader group of 80 university students.

The original experiment had a single person try 8 different cups, but this experiemnt had 80 different people try a single cup and then make their guess.

### Statements of the Null Hypothesis and Alternative Hypothesis

##### Null Hypothesis ($H_0$):

$H_0$: p = 0.5

The proportion of STA130 students who can correctly identify the order in which the milk and tea are poured is equal to 0.5, meaning that their ability is no better than random guessing.


##### Alternative Hypothesis ($H_1$):

$H_1$: p > 0.5
        
The proportion of STA130 students who can correctly identify the order is greater than 0.5, indicating that they have some genuine ability to distinguish the pouring order beyond random chance.

## Quantitative Analysis

We will conduct a formal hypothesis test to assess the validity of $H_0$. The analysis involves calculating the observed proportion of correct identifications and simulating a sampling distribution under the assumption of $H_0$ (random guessing). We will then use the simulation results to compute a one-sided p-value, which quantifies the likelihood of observing a proportion as high as the one observed if $H_0$ were true.

##### Observed Data:
Sample Size (): 80 students.

Correctly Identified The Right Answer: 49 students.

Observed Proportion:

49/80 = 0.6125

### Methodology Code and Explanations

This is the code I used to simulate the hypothesis test and calculate the one-sided p-value:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Given data
sample_size = 80
observed_correct = 49
observed_proportion = observed_correct / sample_size

# Null hypothesis proportion
p_null = 0.5

# Number of simulations
num_simulations = 10000

# Simulate the null distribution using a binomial distribution
np.random.seed(40)  # Set the seed for reproducibility
simulated_proportions = np.random.binomial(n=sample_size, p=p_null, size=num_simulations) / sample_size

# Calculate the p-value for a one-sided test (greater than observed proportion)
p_value = np.mean(simulated_proportions >= observed_proportion)

print(f"Observed proportion: {observed_proportion}")
print(f"Simulated one-tailed p-value: {p_value}")

# Plot the simulated distribution
plt.hist(simulated_proportions, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f'Observed = {observed_proportion}')
plt.title('Simulated Sampling Distribution under H0')
plt.xlabel('Proportion Correct')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### What The Code Does:

The code simulates 10,000 samples of 80 students, assuming a 50% chance of being correct under $H_0$. It calculates the proportion of simulated samples where the proportion of correct identifications is greater than or equal to 0.6125. A histogram then displays the simulated sampling distribution, with the observed proportion marked.

### Supporting Visualizations

The histogram illustrates the distribution of simulated proportions under the null hypothesis.

The red dashed line represents the observed proportion of 0.6125.
And everything to the right of this line represents the proportion of simulations that are as extreme as the observed result, corresponding to the one-sided p-value.
